   $$
   \int c \, dx = cx + C
   $$

   $$

### **2. Compute**: $ \int (x^2 + 2) \, dx $

1. Use the **linearity of integration**:
   $$
   \int (x^2 + 2) \, dx = \int x^2 \, dx + \int 2 \, dx
   $$
2. For $ \int x^2 \, dx $, apply the power rule:
   $$
   \int x^n \, dx = \frac{x^{n+1}}{n+1}, \quad \text{(valid for \( n \neq -1 \))}.
   $$
   Here, $ n = 2 $:
   $$
   \int x^2 \, dx = \frac{x^{2+1}}{2+1} = \frac{x^3}{3}
   $$
3. For $ \int 2 \, dx $, treat $2 $ as a constant:
   $$
   \int 2 \, dx = 2x
   $$
4. Combine the results:
   $$
   \int (x^2 + 2) \, dx = \frac{x^3}{3} + 2x + C
   $$

**Final Answer**:
$$
\int (x^2 + 2) \, dx = \frac{x^3}{3} + 2x + C
$$

### **3. Compute**: $ \int 2\sin(x) \, dx $

1. Factor out the constant $ 2 $:
   $$
   \int 2\sin(x) \, dx = 2 \int \sin(x) \, dx
   $$
2. Recall the integral of $ \sin(x) $ :
   $$
   \int \sin(x) \, dx = -\cos(x)
   $$
3. Substitute:
   $$
   2 \int \sin(x) \, dx = 2 \cdot (-\cos(x)) = -2\cos(x)
   $$
4. Add the constant of integration $ C $:
   $$
   \int 2\sin(x) \, dx = -2\cos(x) + C
   $$

**Final Answer**:
$$
\int 2\sin(x) \, dx = -2\cos(x) + C
$$

### **13. Compute**$ \int x \ln(x) \, dx $

1. Use integration by parts:
   $$
   \int u \, dv = uv - \int v \, du
   $$
   Let $ u = \ln(x) $ and $ dv = x \, dx $. Then:
   $$
   du = \frac{1}{x}dx, \quad v = \frac{x^2}{2}
   $$

2. Apply the formula:
   $$
   \int x \ln(x) \, dx = \frac{x^2}{2} \ln(x) - \int \frac{x^2}{2} \cdot \frac{1}{x} \, dx
   $$

3. Simplify the remaining integral:
   $$
   \int \frac{x^2}{2} \cdot \frac{1}{x} \, dx = \int \frac{x}{2} \, dx = \frac{1}{2} \int x \, dx = \frac{1}{2} \cdot \frac{x^2}{2} = \frac{x^2}{4}
   $$

4. Combine results:
   $$
   \int x \ln(x) \, dx = \frac{x^2}{2} \ln(x) - \frac{x^2}{4} + C
   $$

**Answer**: $ \frac{x^2}{2} \ln(x) - \frac{x^2}{4} + C $

### **14. Compute**$ \int x e^x \, dx $

1. Use integration by parts:
   $$
   \int u \, dv = uv - \int v \, du
   $$
   Let $ u = x $ and $ dv = e^x dx $. Then:
   $$
   du = dx, \quad v = e^x
   $$

2. Apply the formula:
   $$
   \int x e^x \, dx = x e^x - \int e^x \, dx
   $$

3. Solve $ \int e^x \, dx $:
   $$
   \int e^x \, dx = e^x
   $$

4. Combine results:
   $$
   \int x e^x \, dx = x e^x - e^x + C
   $$

**Answer**: $$ x e^x - e^x + C $$


```python
import sympy as sp

# Define the symbol for integration
x = sp.symbols('x')

# Define the integrals
integrals = [
    sp.integrate(1, x),                               # ∫ 1 dx
    sp.integrate(x**2 + 2, x),                       # ∫ (x^2 + 2) dx
    sp.integrate(2 * sp.sin(x), x),                  # ∫ 2 sin(x) dx
    sp.integrate(3/x, x),                            # ∫ 3/x dx
    sp.integrate(1/x**2, x),                         # ∫ 1/x^2 dx
    sp.integrate((1/3) * (x**4 - 5), x),             # ∫ (1/3 * (x^4 - 5)) dx
    sp.integrate(sp.sin(x)**2 + sp.cos(x)**2, x),    # ∫ (sin^2(x) + cos^2(x)) dx
    sp.integrate(5 * sp.sin(x) + 3 * sp.exp(x), x),  # ∫ (5 sin(x) + 3e^x) dx
    sp.integrate(sp.root(x, 3), x),                  # ∫ (x^(1/3)) dx
    sp.integrate(sp.sqrt(10 * x), x),                # ∫ sqrt(10x) dx
    sp.integrate(sp.cos((5/2)*x + 3), x),            # ∫ cos((5/2)x + 3) dx
    sp.integrate(sp.cos(sp.ln(x)) / x, x),           # ∫ cos(ln(x)) / x dx
    sp.integrate(x * sp.ln(x), x),                   # ∫ x ln(x) dx
    sp.integrate(x * sp.exp(x), x)                   # ∫ x e^x dx
]

# Print the results
print("Solutions to the integrals:")
for i, integral in enumerate(integrals, start=1):
    print(f"{i}. {integral}")
```

1. $ \int_{0}^{\pi} (2x + 1) \, dx $
2. $ \int_{0}^{\pi} x^2 \, dx $

---

### Solution

### 1. Compute $ \int_{0}^{\pi} (2x + 1) \, dx $

#### Step 1: Set up the integral
The integral is:
$$
\int_{0}^{\pi} (2x + 1) \, dx
$$

#### Step 2: Compute the indefinite integral
Using the rules of integration:
- $ \int x^n \, dx = \frac{x^{n+1}}{n+1} + C $
- $ \int c \, dx = cx + C $, where $ c $ is a constant.

We integrate $ 2x + 1 $:
$$
\int (2x + 1) \, dx = \int 2x \, dx + \int 1 \, dx = \frac{2x^2}{2} + x + C = x^2 + x + C
$$

#### Step 3: Apply the limits of integration
Using the Fundamental Theorem of Calculus:
$$
\int_{0}^{\pi} (2x + 1) \, dx = \left[ x^2 + x \right]_{0}^{\pi}
$$

Evaluate at the bounds:
$$
\text{At } x = \pi: (\pi^2 + \pi)
$$
$$
\text{At } x = 0: (0^2 + 0) = 0
$$

Subtract:
$$
\int_{0}^{\pi} (2x + 1) \, dx = (\pi^2 + \pi) - 0 = \pi^2 + \pi
$$

---

### 2. Compute $ \int_{0}^{\pi} x^2 \, dx $

#### Step 1: Set up the integral
The integral is:
$$
\int_{0}^{\pi} x^2 \, dx
$$

#### Step 2: Compute the indefinite integral
Using the power rule for integration:
$$
\int x^2 \, dx = \frac{x^{2+1}}{2+1} + C = \frac{x^3}{3} + C
$$

#### Step 3: Apply the limits of integration
Using the Fundamental Theorem of Calculus:
$$
\int_{0}^{\pi} x^2 \, dx = \left[ \frac{x^3}{3} \right]_{0}^{\pi}
$$

Evaluate at the bounds:
$$
\text{At } x = \pi: \frac{\pi^3}{3}
$$
$$
\text{At } x = 0: \frac{0^3}{3} = 0
$$

Subtract:
$$
\int_{0}^{\pi} x^2 \, dx = \frac{\pi^3}{3} - 0 = \frac{\pi^3}{3}
$$

---

## Final Results:
1. $$ \int_{0}^{\pi} (2x + 1) \, dx = \pi^2 + \pi $$
2. $$ \int_{0}^{\pi} x^2 \, dx = \frac{\pi^3}{3} $$

### Problem 3
### Solution for the Area of the Bounded Region

### Problem
Calculate the area of the region bounded by the lines:
- $x = 1$,
- $x = 2$,
- $y = 0$, and
- $y = x^2 + 1$.

---

### Step 1: Set up the integral
The region is bounded by:
- **Top Function**: $y = x^2 + 1$,
- **Bottom Function**: $y = 0$,
- **Interval**: $[x = 1, x = 2]$.

The formula for the area is:
$$
\text{Area} = \int_a^b \text{(Top Function - Bottom Function)} \, dx
$$

Substitute the given functions and limits:
$$
\text{Area} = \int_{1}^{2} \left( (x^2 + 1) - 0 \right) \, dx = \int_{1}^{2} (x^2 + 1) \, dx
$$
---

### Step 2: Compute the integral
Expand the integral:
$$
\int_{1}^{2} (x^2 + 1) \, dx = \int_{1}^{2} x^2 \, dx + \int_{1}^{2} 1 \, dx
$$

### Part 1: $ \int_{1}^{2} x^2 \, dx $
Use the power rule:
$$
\int x^n \, dx = \frac{x^{n+1}}{n+1} + C
$$
$$
\int x^2 \, dx = \frac{x^{3}}{3}
$$

Apply the limits:
$$
\left[ \frac{x^3}{3} \right]_{1}^{2} = \frac{2^3}{3} - \frac{1^3}{3} = \frac{8}{3} - \frac{1}{3} = \frac{7}{3}
$$

### Part 2: $\int_{1}^{2} 1 \, dx $
The integral of a constant $1$ is:
$$
\int 1 \, dx = x
$$

Apply the limits:
$$
\left[ x \right]_{1}^{2} = 2 - 1 = 1
$$

---

### Step 3: Combine the results
Add the results of the two parts:
$$
\text{Area} = \frac{7}{3} + 1 = \frac{7}{3} + \frac{3}{3} = \frac{10}{3}
$$

---

## Final Answer
The area of the region bounded by the given lines is:
$$
\boxed{\frac{10}{3}}
$$

### Calculating the Area Under the Sine Curve $ y = \sin(x) $ Over $[0, \pi]$

---

### Problem
We need to calculate the area under the curve $ y = \sin(x) $ over the interval $[0, \pi]$. This is represented by the definite integral:
$
P = \int_{0}^{\pi} \sin(x) \, dx
$

---

### Step 1: Understand the integral
The definite integral $ \int_{a}^{b} f(x) \, dx $ gives the signed area between the curve $ f(x) $ and the x-axis over the interval $[a, b]$. For $ \sin(x) $, the function is **positive** over $[0, \pi]$, so the integral will represent the actual area under the curve.

---

### Step 2: Write the integral
$$
P = \int_{0}^{\pi} \sin(x) \, dx
$$

---

### Step 3: Compute the indefinite integral
Using standard trigonometric integration rules:
$$
\int \sin(x) \, dx = -\cos(x) + C
$$
Here:
- $ -\cos(x) $ is the antiderivative of $ \sin(x) $.
- $ C $ is the constant of integration, which is omitted for definite integrals.

---

### Step 4: Apply the limits of integration
Using the Fundamental Theorem of Calculus:
$$
P = \left[ -\cos(x) \right]_{0}^{\pi}
This means we evaluate $ -\cos(x) $ at the upper limit $ x = \pi $ and the lower limit $ x = 0 $, and subtract the results.

---

### Step 5: Evaluate at $ x = \pi $
$$
-\cos(\pi) = -(-1) = 1
$$
Here, $ \cos(\pi) = -1 $, and the negative sign makes it $ +1 $.

---

### Step 6: Evaluate at $ x = 0 $
$$
-\cos(0) = -(1) = -1
$
Here, $ \cos(0) = 1 $, and the negative sign makes it $ -1 $.

---

### Step 7: Subtract the values
The definite integral is the difference:
$$
P = \left[ -\cos(x) \right]_{0}^{\pi} = -\cos(\pi) - (-\cos(0))
$$
$$
P = 1 - (-1) = 1 + 1 = 2
$$

---

### Final Answer
The area under the sine curve $ y = \sin(x) $ from $ x = 0 $ to $ x = \pi $ is:
$$
\boxed{2}
$$

---

### Explanation of the Result
1. The sine function $ y = \sin(x) $ starts at $ 0 $ at $ x = 0 $, reaches its maximum value of $ 1 $ at $ x = \frac{\pi}{2} $, and returns to $ 0 $ at $ x = \pi $.
2. Over this interval, the sine function is always **positive**, so the integral represents the total area under the curve.
3. The computation confirms that the total area is $ 2 $.

### Calculating the Arc Length of the Sine Curve $ y = \sin(x) $ Over $[0, \pi]$

---

### **Problem**
We need to calculate the arc length of the sine curve $ y = \sin(x) $ over the interval $[0, \pi]$. The formula for arc length is:
$$
L = \int_a^b \sqrt{1 + \left( \frac{dy}{dx} \right)^2} \, dx
$$

---

### **Step 1: The Arc Length Formula**
The arc length $ L $ is calculated as:
$$
L = \int_a^b \sqrt{1 + \left( f'(x) \right)^2} \, dx
$$
Here:
- $ f(x) = \sin(x) $
- The interval is $ [0, \pi] $.

---

### **Step 2: Compute the Derivative**
The derivative of $ y = \sin(x) $ is:
$$
f'(x) = \frac{dy}{dx} = \cos(x)
$$

---

### **Step 3: Substitute into the Formula**
Substitute $ f'(x) = \cos(x) $ into the arc length formula:
$$
L = \int_0^\pi \sqrt{1 + \cos^2(x)} \, dx
$$

---

### **Step 4: Behavior of $ \sqrt{1 + \cos^2(x)} $**
The term $ \sqrt{1 + \cos^2(x)} $ oscillates because $ \cos^2(x) $ varies between 0 and 1:
1. At $ x = 0 $: $ \cos^2(0) = 1 $, so $ \sqrt{1 + \cos^2(x)} = \sqrt{2} $.
2. At $ x = \frac{\pi}{2} $: $ \cos^2(\pi/2) = 0 $, so $ \sqrt{1 + \cos^2(x)} = 1 $.
3. At $ x = \pi $: $ \cos^2(\pi) = 1 $, so $ \sqrt{1 + \cos^2(x)} = \sqrt{2} $.

Thus, $ \sqrt{1 + \cos^2(x)} $ oscillates between $ 1 $ and $ \sqrt{2} $.

---

### **Step 5: Simplify the Integral**
The integral becomes:
$$
L = \int_0^\pi \sqrt{1 + \cos^2(x)} \, dx
$$
This integral cannot be solved using elementary functions because $ \sqrt{1 + \cos^2(x)} $ does not have a closed-form antiderivative.

---

### **Step 6: Numerical Evaluation**
Since $ \sqrt{1 + \cos^2(x)} $ has no elementary antiderivative, the integral must be evaluated numerically. Using numerical integration methods or a symbolic computation tool, we find:
$$
L = \int_0^\pi \sqrt{1 + \cos^2(x)} \, dx \approx 3.82
$$

---

### **Step 7: Intuition Behind the Result**
1. **Geometric Meaning**: The arc length measures the total "distance traveled" along the curve, taking into account both horizontal and vertical changes.
   - For $ y = \sin(x) $, the curve oscillates smoothly between $ y = 0 $ and $ y = 1 $, adding to the total length.

2. **Comparison to the Interval**: The interval $[0, \pi]$ has a straight-line length of $ \pi \approx 3.14 $. The arc length $ L \approx 3.82 $ is slightly larger because the curve has peaks and valleys, increasing the total distance.

---

### **Final Answer**
The arc length of the sine curve $ y = \sin(x) $ over $[0, \pi]$ is approximately:
$$
\boxed{3.82}
$$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Define the sine function and the integrand for arc length
x = np.linspace(0, np.pi, 500)
y = np.sin(x)

# Arc length integrand: sqrt(1 + cos^2(x))
def arc_length_integrand(x):
    return np.sqrt(1 + np.cos(x)**2)

# Compute the arc length numerically
arc_length, _ = quad(arc_length_integrand, 0, np.pi)

# Plot the sine curve
plt.figure(figsize=(10, 6))
plt.plot(x, y, label="$y = \sin(x)$", color="blue")

# Fill the area under the curve for visual effect
plt.fill_between(x, y, color="lightblue", alpha=0.5, label=f"Arc Length ≈ {arc_length:.2f}")

# Add titles, labels, and legend
plt.title("Arc Length of $y = \sin(x)$ Over $[0, \pi]$", fontsize=16)
plt.xlabel("$x$", fontsize=14)
plt.ylabel("$y$", fontsize=14)
plt.axhline(0, color="black", linewidth=0.8)
plt.legend(fontsize=12)
plt.grid(alpha=0.3)

# Display the plot
plt.show()
```

### Total Distance Traveled by a Particle Between $ t = 0 $ and $ t = 2 $

---

### **Problem**
The position of a particle is given by:
$$
x(t) = 3t^2 - 6t + 1
$$
We are tasked with finding the **total distance traveled** by the particle between $ t = 0 $ and $ t = 2 $.

---

### **Step 1: Understanding Total Distance**
The **total distance** traveled is calculated using the absolute value of the velocity:
$$
\text{Distance} = \int_{0}^{2} |v(t)| \, dt
$$
where $ v(t) = \frac{dx}{dt} $ is the velocity of the particle.

---

### **Step 2: Compute the Velocity Function**
The velocity $ v(t) $ is the derivative of the position function $ x(t) $:
$$
x(t) = 3t^2 - 6t + 1
$$
$$
v(t) = \frac{dx}{dt} = 6t - 6
$$

---

### **Step 3: Find When the Particle Changes Direction**
The particle changes direction when $ v(t) = 0 $:
$$
6t - 6 = 0
$$
$$
t = 1
$$

This means the particle changes direction at $ t = 1 $. To calculate the total distance, we break the integral into two parts:
1. From $ t = 0 $ to $ t = 1 $, where $ v(t) < 0 $ (the particle moves backward).
2. From $ t = 1 $ to $ t = 2 $, where $ v(t) > 0 $ (the particle moves forward).

---

### **Step 4: Set Up the Total Distance Integral**
The total distance is:
$$
\text{Distance} = \int_{0}^{1} |v(t)| \, dt + \int_{1}^{2} |v(t)| \, dt
$$

Since $ v(t) = 6t - 6 $:
- For $ 0 \leq t \leq 1 $, $ v(t) < 0 $, so $ |v(t)| = -(6t - 6) = -6t + 6 $.
- For $ 1 \leq t \leq 2 $, $ v(t) > 0 $, so $ |v(t)| = 6t - 6 $.

Thus:
$$
\text{Distance} = \int_{0}^{1} (-6t + 6) \, dt + \int_{1}^{2} (6t - 6) \, dt
$$

---

### **Step 5: Compute Each Integral**

#### **First Integral**: $ \int_{0}^{1} (-6t + 6) \, dt $
$$
\int (-6t + 6) \, dt = \int -6t \, dt + \int 6 \, dt
$$
$$
= -3t^2 + 6t + C
$$

Evaluate at $ t = 0 $ and $ t = 1 $:
$$
\left[ -3t^2 + 6t \right]_{0}^{1} = \left( -3(1)^2 + 6(1) \right) - \left( -3(0)^2 + 6(0) \right)
$$
$$
= (-3 + 6) - (0) = 3
$$

---

#### **Second Integral**: $ \int_{1}^{2} (6t - 6) \, dt $
$$
\int (6t - 6) \, dt = \int 6t \, dt - \int 6 \, dt
$$
$$
= 3t^2 - 6t + C
$$

Evaluate at $ t = 1 $ and $ t = 2 $:
$$
\left[ 3t^2 - 6t \right]_{1}^{2} = \left( 3(2)^2 - 6(2) \right) - \left( 3(1)^2 - 6(1) \right)
$$
$$
= \left( 12 - 12 \right) - \left( 3 - 6 \right)
$$
$$
= 0 - (-3) = 3
$$

---

### **Step 6: Add the Results**
The total distance traveled is:
$$
\text{Distance} = \int_{0}^{1} (-6t + 6) \, dt + \int_{1}^{2} (6t - 6) \, dt
$$
$$
\text{Distance} = 3 + 3 = 6
$$

---

### **Final Answer**
The total distance traveled by the particle between $ t = 0 $ and $ t = 2 $ is:
$$
\boxed{6}
$$

---

### **Explanation**
1. The particle moves backward from $ t = 0 $ to $ t = 1 $, traveling a distance of $ 3 $.
2. The particle moves forward from $ t = 1 $ to $ t = 2 $, traveling another distance of $ 3 $.
3. The total distance is the sum of these two distances: $ 3 + 3 = 6 $.
This accounts for the absolute value of the velocity at all times.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Define the velocity function (derivative of x(t))
def velocity(t):
    return 6*t - 6

# Define the absolute value of velocity for total distance calculation
def abs_velocity(t):
    return np.abs(velocity(t))

# Compute total distance traveled using numerical integration
distance, _ = quad(abs_velocity, 0, 2)

# Generate t values for plotting
t = np.linspace(0, 2, 500)
position = 3*t**2 - 6*t + 1
vel = velocity(t)

# Plot position function
plt.figure(figsize=(12, 6))

# Position curve
plt.subplot(1, 2, 1)
plt.plot(t, position, label="$x(t) = 3t^2 - 6t + 1$", color="blue")
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.title("Position Function $x(t)$", fontsize=14)
plt.xlabel("$t$", fontsize=12)
plt.ylabel("$x(t)$", fontsize=12)
plt.grid(alpha=0.3)
plt.legend(fontsize=12)

# Velocity curve
plt.subplot(1, 2, 2)
plt.plot(t, vel, label="$v(t) = 6t - 6$", color="green")
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.title("Velocity Function $v(t)$", fontsize=14)
plt.xlabel("$t$", fontsize=12)
plt.ylabel("$v(t)$", fontsize=12)
plt.grid(alpha=0.3)
plt.legend(fontsize=12)

# Display total distance traveled
plt.suptitle(f"Total Distance Traveled: {distance:.2f}", fontsize=16)
plt.tight_layout()
plt.subplots_adjust(top=0.85)
plt.show()
```
