### **1. Compute**: $ \int 1 \, dx $

1. The integral of a constant $ c $ is given by:
   $$
   \int c \, dx = cx + C
   $$
   where $ C $ is the constant of integration.
2. Here, $ c = 1 $:
   $$
   \int 1 \, dx = 1 \cdot x + C
   $$

**Final Answer**:
$$
\int 1 \, dx = x + C
$$

### **2. Compute**: $ \int (x^2 + 2) \, dx $

1. Use the **linearity of integration**:
   $$
   \int (x^2 + 2) \, dx = \int x^2 \, dx + \int 2 \, dx
   $$
2. For $ \int x^2 \, dx $, apply the power rule:
   $$
   \int x^n \, dx = \frac{x^{n+1}}{n+1}, \quad \text{(valid for \( n \neq -1 \))}.
   $$
   Here, $ n = 2 $:
   $$
   \int x^2 \, dx = \frac{x^{2+1}}{2+1} = \frac{x^3}{3}
   $$
3. For $ \int 2 \, dx $, treat $2 $ as a constant:
   $$
   \int 2 \, dx = 2x
   $$
4. Combine the results:
   $$
   \int (x^2 + 2) \, dx = \frac{x^3}{3} + 2x + C
   $$

**Final Answer**:
$$
\int (x^2 + 2) \, dx = \frac{x^3}{3} + 2x + C
$$

### **3. Compute**: $ \int 2\sin(x) \, dx $

1. Factor out the constant $ 2 $:
   $$
   \int 2\sin(x) \, dx = 2 \int \sin(x) \, dx
   $$
2. Recall the integral of $ \sin(x) $ :
   $$
   \int \sin(x) \, dx = -\cos(x)
   $$
3. Substitute:
   $$
   2 \int \sin(x) \, dx = 2 \cdot (-\cos(x)) = -2\cos(x)
   $$
4. Add the constant of integration $ C $:
   $$
   \int 2\sin(x) \, dx = -2\cos(x) + C
   $$

**Final Answer**:
$$
\int 2\sin(x) \, dx = -2\cos(x) + C
$$

### **4. Compute**: $ \int \frac{3}{x} \, dx $

1. Factor out the constant $ 3 $:
   $$
   \int \frac{3}{x} \, dx = 3 \int \frac{1}{x} \, dx
   $$
    2. Recall the integral of $ \frac{1}{x} $:
   $$
   \int \frac{1}{x} \, dx = \ln|x|
   $$
3. Substitute:
   $$
   3 \int \frac{1}{x} \, dx = 3 \cdot \ln|x| = 3\ln|x|
   $$
4. Add the constant of integration $ C $:
   $$
   \int \frac{3}{x} \, dx = 3\ln|x| + C
   $$

**Final Answer**:
$$
\int \frac{3}{x} \, dx = 3\ln|x| + C
$$

### **5. Compute**: $ \int \frac{1}{x^2} \, dx $

1. Rewrite $ \frac{1}{x^2} $ as $ x^{-2} $:
   $$
   \int \frac{1}{x^2} \, dx = \int x^{-2} \, dx
   $$
2. Use the power rule:
   $$
   \int x^n \, dx = \frac{x^{n+1}}{n+1}, \quad \text{(valid for \( n \neq -1 \))}.
   $$
   Here, $ n = -2 $:
   $$
   \int x^{-2} \, dx = \frac{x^{-2+1}}{-2+1} = \frac{x^{-1}}{-1} = -\frac{1}{x}
   $$
3. Add the constant of integration $ C $:
   $$
   \int \frac{1}{x^2} \, dx = -\frac{1}{x} + C
   $$

**Final Answer**:
$$
\int \frac{1}{x^2} \, dx = -\frac{1}{x} + C
$$

### **6. Compute**: $ \int \left( \frac{1}{3}x^4 - 5 \right) \, dx $

1. Split the integral:
   $$
   \int \left( \frac{1}{3}x^4 - 5 \right) \, dx = \int \frac{1}{3}x^4 \, dx - \int 5 \, dx
   $$
2. For $ \int \frac{1}{3}x^4 \, dx $, factor out $ \frac{1}{3} $ and apply the power rule:
   $$
   \int \frac{1}{3}x^4 \, dx = \frac{1}{3} \cdot \frac{x^{4+1}}{4+1} = \frac{x^5}{15}
   $$
3. For $ \int 5 \, dx $, treat $ 5 $ as a constant:
   $$
   \int 5 \, dx = 5x
   $$
4. Combine the results:
   $$
   \int \left( \frac{1}{3}x^4 - 5 \right) \, dx = \frac{x^5}{15} - 5x + C
   $$

**Final Answer**:
$$
\int \left( \frac{1}{3}x^4 - 5 \right) \, dx = \frac{x^5}{15} - 5x + C
$$

### **7. Compute**: $\int (\sin^2(x) + \cos^2(x)) \, dx$

1. Recall the **Pythagorean identity**:
   $$
   \sin^2(x) + \cos^2(x) = 1
   $$
2. Substitute:
   $$
   \int (\sin^2(x) + \cos^2(x)) \, dx = \int 1 \, dx
   $$
3. Integrate $1$:
   $
   \int 1 \, dx = x
   $
4. Add the constant of integration $C$:
   $$
   \int (\sin^2(x) + \cos^2(x)) \, dx = x + C
   $$

**Final Answer**:
$$
\int (\sin^2(x) + \cos^2(x)) \, dx = x + C
$$
