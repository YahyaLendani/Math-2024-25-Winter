```python
import sympy as sp

# Define the variable
x = sp.Symbol('x')
t = sp.Symbol('t')

# Part 1: Compute indefinite integrals
integrals = {
    "1": sp.integrate(1, x),
    "2": sp.integrate(x**2 + 2, x),
    "3": sp.integrate(2 * sp.sin(x), x),
    "4": sp.integrate(3 / x, x),
    "5": sp.integrate(1 / x**2, x),
    "6": sp.integrate((1/3) * x**4 - 5, x),
    "7": sp.integrate(sp.sin(x)**2 + sp.cos(x)**2, x),
    "8": sp.integrate(5 * sp.sin(x) + 3 * sp.exp(x), x),
    "9": sp.integrate(sp.cbrt(x), x),
    "10": sp.integrate(sp.sqrt(10 * x), x),
    "11": sp.integrate(sp.cos(5/2 * x + 3), x),
    "12": sp.integrate(sp.cos(sp.ln(x)) / x, x),
    "13": sp.integrate(x * sp.ln(x), x),
    "14": sp.integrate(x * sp.exp(x), x),
}

# Part 2: Definite integrals over [0, π]
f1 = 2 * x + 1
g1 = x**2
integral_f1 = sp.integrate(f1, (x, 0, sp.pi))
integral_g1 = sp.integrate(g1, (x, 0, sp.pi))

# Part 3: Calculate the area of the region bounded by the lines
area_bounded = sp.integrate((x**2 + 1), (x, 1, 2))

# Part 4: Area under the sine curve over [0, π]
area_sine_curve = sp.integrate(sp.sin(x), (x, 0, sp.pi))

# Part 5: Calculate the length of the sine curve over [0, π]
f = sp.sin(x)
f_prime = sp.diff(f, x)
length_sine_curve = sp.integrate(sp.sqrt(1 + f_prime**2), (x, 0, sp.pi))

# Part 6: Find the distance of the moving particle between t=0 and t=2
x_t = 3 * t**2 - 6 * t + 1
velocity = sp.diff(x_t, t)
distance = sp.integrate(sp.sqrt(velocity**2), (t, 0, 2))

# Print results
print("Part 1: Indefinite Integrals")
for key, integral in integrals.items():
    print(f"Integral {key}: {integral}")

print("\nPart 2: Definite Integrals over [0, π]")
print(f"∫[0,π] (2x + 1) dx = {integral_f1}")
print(f"∫[0,π] (x^2) dx = {integral_g1}")

print("\nPart 3: Area of the Region Bounded by the Lines")
print(f"Area = {area_bounded}")

print("\nPart 4: Area Under the Sine Curve Over [0, π]")
print(f"Area = {area_sine_curve}")

print("Part 5: Length of the Sine Curve over [0, π]")
print(f"L = {length_sine_curve}")

print("\nPart 6: Distance of the Moving Particle from t=0 to t=2")
print(f"Distance = {distance}")
```
