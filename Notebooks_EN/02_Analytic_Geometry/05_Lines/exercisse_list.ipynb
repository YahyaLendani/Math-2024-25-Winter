## **1. The line passes through points $ A(1, 2) $ and $ B(3, 4) $. Find the equation of the line.**

### **Step 1: Find the slope of the line**

The slope $ m $ of a line passing through two points $ A(x_1, y_1) $ and $ B(x_2, y_2) $ is given by the formula:

$$
m = \frac{y_2 - y_1}{x_2 - x_1}
$$

Substituting the values of points $ A(1, 2) $ and $ B(3, 4) $:

$$
m = \frac{4 - 2}{3 - 1} = \frac{2}{2} = 1
$$

### **Step 2: Use the point-slope form of the line equation**

The point-slope form of a line is:

$$
y - y_1 = m(x - x_1)
$$

Using point $ A(1, 2) $ and the slope $ m = 1 $:

$$
y - 2 = 1(x - 1)
$$

Simplifying the equation:

$$
y - 2 = x - 1 \quad \Rightarrow \quad y = x + 1
$$

Thus, the equation of the line is:

$$
y = x + 1
$$

## **2. The line passes through point $ A(1, 2) $ and is parallel to the line $ y = 2x + 3 $. Find the equation of the line.**

### **Step 1: Identify the slope of the parallel line**

Since the line is parallel to $ y = 2x + 3 $, it has the same slope. The slope of the line $ y = 2x + 3 $ is $ m = 2 $.

### **Step 2: Use the point-slope form of the line equation**

The point-slope form is:

$$
y - y_1 = m(x - x_1)
$$

Substitute $ m = 2 $ and point $ A(1, 2) $:

$$
y - 2 = 2(x - 1)
$$

Simplifying the equation:

$$
y - 2 = 2x - 2 \quad \Rightarrow \quad y = 2x
$$

Thus, the equation of the line is:

$$
y = 2x
$$

## **3. The line passes through point $ A(1, 2) $ and is perpendicular to the line $ y = 2x + 3 $. Find the equation of the line.**

### **Step 1: Find the slope of the perpendicular line**

The slope of the line $ y = 2x + 3 $ is $ m = 2 $. The slope of the line perpendicular to this line is the negative reciprocal:

$$
m_{\text{perpendicular}} = -\frac{1}{2}
$$

### **Step 2: Use the point-slope form of the line equation**

Using point $ A(1, 2) $ and the slope $ m = -\frac{1}{2} $:

$$
y - 2 = -\frac{1}{2}(x - 1)
$$

Simplifying the equation:

$$
y - 2 = -\frac{1}{2}x + \frac{1}{2} \quad \Rightarrow \quad y = -\frac{1}{2}x + \frac{5}{2}
$$

Thus, the equation of the line is:

$$
y = -\frac{1}{2}x + \frac{5}{2}
$$

## **4. We have two lines $ y = 2x + 3 $ and $ y = 3x + 2 $. Find the intersection point of these lines and calculate the angle between them.**

### **Step 1: Find the intersection point**

Set the equations equal to each other to find the intersection point:

$$
2x + 3 = 3x + 2
$$

Solving for $ x $:

$$
2x - 3x = 2 - 3 \quad \Rightarrow \quad -x = -1 \quad \Rightarrow \quad x = 1
$$

Substitute $ x = 1 $ into either equation (use $ y = 2x + 3 $):

$$
y = 2(1) + 3 = 5
$$

Thus, the intersection point is $ (1, 5) $.

### **Step 2: Calculate the angle between the two lines**

The formula for the angle $ \theta $ between two lines with slopes $ m_1 $ and $ m_2 $ is:

$$
\tan \theta = \left| \frac{m_1 - m_2}{1 + m_1 m_2} \right|
$$

For the lines $ y = 2x + 3 $ and $ y = 3x + 2 $, the slopes are $ m_1 = 2 $ and $ m_2 = 3 $. Substituting into the formula:

$$
\tan \theta = \left| \frac{2 - 3}{1 + 2 \times 3} \right| = \left| \frac{-1}{7} \right| = \frac{1}{7}
$$

Thus, the angle is:

$$
\theta = \tan^{-1}\left(\frac{1}{7}\right) \approx 8.13^\circ
$$

## **5. Write the equation of the line passing through point $ A(1, 2) $ and parallel to the vector $ \mathbf{v} = [2, 3] $.**

### **Step 1: Identify the slope of the line**

The slope of the line is the same as the direction of the vector $ \mathbf{v} = [2, 3] $. The slope is:

$$
m = \frac{3}{2}
$$

### **Step 2: Use the point-slope form of the line equation**

Using point $ A(1, 2) $ and the slope $ m = \frac{3}{2} $:

$$
y - 2 = \frac{3}{2}(x - 1)
$$

Simplifying the equation:

$$
y - 2 = \frac{3}{2}x - \frac{3}{2} \quad \Rightarrow \quad y = \frac{3}{2}x - \frac{1}{2}
$$

Thus, the equation of the line is:

$$
y = \frac{3}{2}x - \frac{1}{2}
$$

## **6. We have the line $ y = 2x + 3 $. Find an example of a line perpendicular and parallel to it.**

- A line **parallel** to $ y = 2x + 3 $ will have the same slope $ m = 2 $. For example, $ y = 2x - 1 $.
- A line **perpendicular** to $ y = 2x + 3 $ will have a slope that is the negative reciprocal of $ 2 $, which is $ m = -\frac{1}{2} $. For example, $ y = -\frac{1}{2}x + 4 $.

## **7. We have the line $ y = 2x + 3 $ and point $ A(1, 2) $. Find the distance from point $ A $ to the line.**

The formula for the distance $ d $ from a point $ (x_1, y_1) $ to a line $ Ax + By + C = 0 $ is:

$$
d = \frac{|Ax_1 + By_1 + C|}{\sqrt{A^2 + B^2}}
$$

The equation of the line is $ y = 2x + 3 $, which can be written as:

$$
2x - y + 3 = 0
$$

Substitute $ A(1, 2) $ into the formula:

$$
d = \frac{|2(1) - (2) + 3|}{\sqrt{2^2 + (-1)^2}} = \frac{|2 - 2 + 3|}{\sqrt{4 + 1}} = \frac{3}{\sqrt{5}} \approx 1.34
$$

Thus, the distance from point $ A $ to the line is approximately **1.34 units**.

## **8. The line intersects the coordinate axes at points $ A(2, 0) $ and $ B(0, 3) $. Find the equation of the line.**

### **Step 1: Use the two points to find the slope**

The slope $ m $ of the line passing through $ A(2, 0) $ and $ B(0, 3) $ is:

$$
m = \frac{3 - 0}{0 - 2} = \frac{3}{-2} = -\frac{3}{2}
$$

### **Step 2: Use the point-slope form**

Using point $ A(2, 0) $ and slope $ m = -\frac{3}{2} $:

$$
y - 0 = -\frac{3}{2}(x - 2)
$$

Simplifying the equation:

$$
y = -\frac{3}{2}(x - 2) = -\frac{3}{2}x + 3
$$

Thus, the equation of the line is:

$$
y = -\frac{3}{2}x + 3
$$

## **9. Calculate the angle between the line $ y = x + 3 $ and the $ OX $ axis.**

The slope of the line $ y = x + 3 $ is $ m = 1 $. The angle $ \theta $ between the line and the $ OX $ axis is given by:

$$
\tan \theta = |m|
$$

Thus,

$$
\tan \theta = 1
$$

So,

$$
\theta = \tan^{-1}(1) = 45^\circ
$$

## **10. Provide a vector perpendicular to the line $ x + y + 1 = 0 $.**

The equation of the line is $ x + y + 1 = 0 $ , which can be written as $ x + y = -1 $.

The vector perpendicular to this line will be parallel to the normal vector of the line. The normal vector to the line $ x + y = -1 $ is $ \mathbf{n} = [1, 1] $.

Thus, a vector perpendicular to the line is $ \mathbf{v} = [1, 1] $.


```python
import sympy as sp

# Define symbols
x, y = sp.symbols('x y')

# Define the line equation
line_eq = sp.Eq(y, 2 * x + 3)

# Define the point
point_A = (1, 2)

# Rearrange the line equation to the standard form Ax + By + C = 0
line_eq_standard = sp.Eq(2 * x - y + 3, 0)

# Extract coefficients A, B, and C
A = line_eq_standard.lhs.coeff(x)
B = line_eq_standard.lhs.coeff(y)
C = line_eq_standard.lhs.as_expr().subs({x: 0, y: 0})

# Calculate the distance using the formula
distance = sp.Abs(A * point_A[0] + B * point_A[1] + C) / sp.sqrt(A**2 + B**2)

# Display the result
print(f"Distance from point {point_A} to the line {line_eq}: {distance}")
```
